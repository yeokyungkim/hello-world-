```
## Homework 4 - from the rating column count the % of WORST DISGUSTING movies in 2020 compare to 2008 Netflix Movies and TV Shows

## Adding the csv file to the pandas library
import pandas as pd
##Importing the dataset to a dataframe
film_list = pd.read_csv("netflix_titles.csv")

## Calculate total number of Netflix film list by type
number_of_film_list= film_list.type.count()
#print(number_of_film_list)

## Slicing: disply movie list from the Netflix film list and extract the data by type and release year
Movie_list = film_list[film_list['type']=="Movie"][['type', 'release_year','rating']]
#display(Movie_list)



## Slicing: display movie list produced in 2020
Movie_list_produced_in_2020 = Movie_list[Movie_list['release_year']==2020][['type', 'release_year','rating']]
#display(Movie_list_produced_in_2020)

## Slicing: display movie list produced in 2020 with R rating only
Movie_list_produced_in_2020_R_Rating = Movie_list_produced_in_2020[Movie_list_produced_in_2020['rating']=="R"][['type', 'release_year','rating']]
#display(Movie_list_produced_in_2020_R_Rating)

## Calculate number of movies produced in 2020 with R rating
number_of_R_rating_Movies_produced_in_2020 = Movie_list_produced_in_2020_R_Rating.type.count()
#print(number_of_R_rating_Movies_produced_in_2020)

## Slicing: display Netflix TV Shows and Movies produced in 2008
Netflix_TVshows_Movies_produced_in_2008 = film_list[film_list['release_year']==2008][['type', 'release_year','rating']]
#display(Netflix_TVshows_Movies_produced_in_2008)

## Slicing: display Netflix TV Shows and Movies produced in 2008 with R rating
Netflix_TVshows_Movies_produced_in_2008_R_Rating = Netflix_TVshows_Movies_produced_in_2008[Netflix_TVshows_Movies_produced_in_2008['rating']=="R"][['type', 'release_year','rating']]
#display(Netflix_TVshows_Movies_produced_in_2008_R_Rating)

## Calculate number of TV shows and movies produced in 2008 with R rating
number_of_R_rating_Netflix_TVshows_Movies_produced_in_2008 = Netflix_TVshows_Movies_produced_in_2008_R_Rating.type.count()
#print(number_of_R_rating_Netflix_TVshows_Movies_produced_in_2008)

Percentage_of_R_rating_Movies_produced_in_2020= number_of_R_rating_Movies_produced_in_2020 / number_of_film_list*100
print("Percentage of R rating movies produced in 2020 is " + str(Percentage_of_R_rating_Movies_produced_in_2020) + " %.")

Percentage_of_R_rating_TVshows_Movies_produced_in_2008= number_of_R_rating_Netflix_TVshows_Movies_produced_in_2008 / number_of_film_list*100
print("Percentage of R rating TV shows and movies produced in 2008 is " + str(Percentage_of_R_rating_TVshows_Movies_produced_in_2008) + " %.")
```
